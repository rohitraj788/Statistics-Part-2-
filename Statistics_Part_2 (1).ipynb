1. What is the purpose of hypothesis testing in statistics, and how is it used to make decisions?

Hypothesis testing is a statistical method used to make decisions about a population based on sample data. It helps determine whether there is enough evidence to support a specific claim (hypothesis) about a parameter (like the mean or proportion).

2. What do we mean by the null hypothesis, and how does it contrast with the alternative hypothesis?

- Null Hypothesis (H₀): Assumes no effect or no difference. It’s the default assumption.

- Alternative Hypothesis (H₁ or Ha): Represents what we aim to support; it indicates the presence of an effect or difference.
You test data to see if you have enough evidence to reject the null hypothesis in favor of the alternative.

3. What is a significance level (α), and why is it critical in hypothesis testing?

The significance level (α) is the threshold for rejecting the null hypothesis. Common values are 0.05 or 0.01. It represents the probability of making a Type I error (i.e., rejecting a true null hypothesis). Lower α means stricter testing.

4. What does the P-value tell us in the context of hypothesis testing?

The P-value is the probability of obtaining a result as extreme as or more extreme than the observed result, assuming the null hypothesis is true. It helps you determine whether to reject the null hypothesis.

5. How should we interpret the P-value when making a decision about a hypothesis?

- If P ≤ α → Reject the null hypothesis (evidence supports the alternative).

- If P > α → Fail to reject the null hypothesis (insufficient evidence).

6. What are Type I and Type II errors, and how do they impact the conclusions we draw?

Type I Error (α): Rejecting a true null hypothesis (false positive).

Type II Error (β): Failing to reject a false null hypothesis (false negative).
You balance these by choosing appropriate sample sizes and significance levels.

7. How do one-tailed and two-tailed hypothesis tests differ, and when should each be used?

- One-tailed test: Tests for a difference in a specific direction (e.g., greater than).

- Two-tailed test: Tests for a difference in both directions (e.g., not equal).
Use one-tailed only when the direction of the effect is known and justified.

8. What is a Z-test, and in what situations is it appropriate to use it?

A Z-test is used when:

- The population variance is known.

- The sample size is large (n ≥ 30).
It compares the sample mean to the population mean using the Z-distribution.

10. What is a T-distribution, and why is it used when sample sizes are small?

The T-distribution is similar to the normal distribution but has heavier tails. It's used:

- When the sample size is small (n < 30)

- When the population standard deviation is unknown
It accounts for the additional uncertainty in estimating the population standard deviation.



11. How do Z-tests and T-tests differ, and how do you decide which to use?

- Z-test: Known population standard deviation and/or large sample size.

- T-test: Unknown population standard deviation and small sample size.
T-test is more conservative and preferred when in doubt.

12. What is a T-test, and how is it applied to compare means in different scenarios?

A T-test is used to compare sample means. Types:

- One-sample T-test: Compare sample mean to known value.

- Two-sample T-test: Compare means of two independent groups.

- Paired T-test: Compare means from the same group at different times (before-after)

13. How are Z-tests and T-tests related in terms of logic and application?

Both test whether a sample mean significantly differs from a known or assumed value. T-tests are used in place of Z-tests when sample sizes are small and population variance is unknown. As sample size increases, the T-distribution approaches the normal distribution.

14. What is a confidence interval, and how can it help us understand the uncertainty in our estimates?

A confidence interval is a range of values within which we expect the population parameter to lie, with a certain level of confidence (e.g., 95%).
It provides a range estimate rather than a single point.

15. What does the margin of error represent, and how does it influence the width of a confidence interval?

The margin of error is the range above and below the sample estimate.
Larger margin → wider confidence interval → more uncertainty
Smaller margin → narrower interval → more precise estimate

16. How is Bayes' Theorem applied in statistics, and what insights does it offer for decision-making?

Bayes’ Theorem updates the probability of a hypothesis based on new evidence. It’s used in:

- Medical testing (diagnosis)

- Spam filtering

- Predictive modeling

19. What is the F-distribution, and when is it used in statistical testing?

The F-distribution is used to compare two variances. It's right-skewed and used in:

- ANOVA

- F-tests
It helps test whether group variances are significantly different.

20. What is an ANOVA test, and what assumptions must be met for its valid use?

ANOVA (Analysis of Variance) tests whether there are significant differences between the means of 3 or more groups.
Assumptions:

- Normal distribution

- Equal variances (homoscedasticity)

- Independent observations

21. What are the various types of ANOVA tests, and how do they differ (e.g., one-way vs. two-way)?

- One-way ANOVA: One independent variable (factor)

- Two-way ANOVA: Two independent variables

- Repeated Measures ANOVA: Same subjects tested under different conditions or times

22. What is the F-test in statistics, and how does it relate to testing variability across groups?

The F-test is used to compare two or more variances. In ANOVA, it tests whether the variability between group means is greater than the variability within the groups.

 1. Perform a Z-test to compare sample mean to population mean

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample = np.random.normal(loc=72, scale=5, size=50)
population_mean = 70
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)

# Z-test
z_stat = (sample_mean - population_mean) / (sample_std / np.sqrt(len(sample)))
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Z-statistic = {z_stat:.2f}, P-value = {p_value:.4f}")

if p_value < 0.05:
    print("Reject null hypothesis: significant difference.")
else:
    print("Fail to reject null: no significant difference.")
```

 3. One-sample Z-test: Sample mean vs Population mean

```python
def one_sample_z_test(data, pop_mean):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    z = (sample_mean - pop_mean) / (sample_std / np.sqrt(len(data)))
    p = 2 * (1 - norm.cdf(abs(z)))
    return z, p

data = np.random.normal(50, 5, 100)
z_stat, p_val = one_sample_z_test(data, 52)
print("Z-statistic:", z_stat)
print("P-value:", p_val)
```

5. Type I and Type II Error Visualization

```python
def visualize_type_errors():
    x = np.linspace(-4, 4, 1000)
    null = norm.pdf(x, 0, 1)
    alt = norm.pdf(x, 1.5, 1)

    plt.plot(x, null, label='H0 (mean=0)', color='blue')
    plt.plot(x, alt, label='H1 (mean=1.5)', color='orange')

    crit = norm.ppf(0.95)
    plt.fill_between(x, null, where=(x >= crit), color='red', alpha=0.3, label='Type I Error')
    plt.fill_between(x, alt, where=(x < crit), color='green', alpha=0.3, label='Type II Error')

    plt.axvline(crit, color='black', linestyle='--')
    plt.title("Type I and Type II Errors")
    plt.legend()
    plt.grid(True)
    plt.show()

visualize_type_errors()
```

9. Confidence Interval for Sample Mean

```python
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    sem = stats.sem(data)
    ci = stats.t.interval(confidence, len(data)-1, loc=mean, scale=sem)
    return ci

data = np.random.normal(70, 10, 100)
print("Confidence Interval:", confidence_interval(data))
```

10. Margin of Error Calculation

```python
conf_level = 0.95
z_star = norm.ppf(1 - (1 - conf_level) / 2)
sample = np.random.normal(80, 12, 100)
sem = np.std(sample, ddof=1) / np.sqrt(len(sample))
margin_of_error = z_star * sem
print("Margin of Error:", margin_of_error)
```

11. Bayesian Inference using Bayes' Theorem

```python
# P(A|B) = [P(B|A) * P(A)] / P(B)

def bayes_theorem(p_a, p_b_given_a, p_b_given_not_a):
    p_not_a = 1 - p_a
    p_b = p_b_given_a * p_a + p_b_given_not_a * p_not_a
    p_a_given_b = (p_b_given_a * p_a) / p_b
    return p_a_given_b

# Example
p_a = 0.01          # Probability of disease
p_b_given_a = 0.99  # True positive rate
p_b_given_not_a = 0.05  # False positive rate

posterior = bayes_theorem(p_a, p_b_given_a, p_b_given_not_a)
print("Posterior Probability (P(Disease | Positive Test)):", posterior)
```
